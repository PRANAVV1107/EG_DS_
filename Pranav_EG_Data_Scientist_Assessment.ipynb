### Running a PCA

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


features = df.drop(['LeagueIndex', 'GameID'], axis=1)  # Exclude the target variable column if present

# Standardize the features (mean=0, variance=1)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply PCA
pca = PCA()
principal_components = pca.fit_transform(scaled_features)

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Cumulative explained variance
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Select the desired number of principal components
n_components = 3 
selected_components = principal_components[:, :n_components]

# Create a DataFrame for the selected components
selected_components_df = pd.DataFrame(selected_components, columns=['PC1', 'PC2', 'PC3'])

if 'LeagueIndex' in df.columns:
    selected_components_df['LeagueIndex'] = df['LeagueIndex']

```

### Plotting a correlation circle of variables

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


# Extract the feature columns from the DataFrame
features = df.drop(['LeagueIndex', 'GameID'], axis=1) # Exclude the target variable column if present

# Standardize the features (mean=0, variance=1)
scaled_features = (features - features.mean()) / features.std()

# Apply PCA
pca = PCA()
principal_components = pca.fit_transform(scaled_features)

# Calculate the correlation matrix of the original features
correlation_matrix = np.corrcoef(scaled_features.values.T)

# Get the eigenvalues and eigenvectors of the correlation matrix
eigenvalues, eigenvectors = np.linalg.eig(correlation_matrix)

# Get the square root of eigenvalues as the singular values
singular_values = np.sqrt(eigenvalues)

# Calculate the correlation circle coordinates
correlation_circle = eigenvectors.T.dot(np.diag(singular_values))

# Set up the plot
fig, ax = plt.subplots(figsize=(15,8))
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.axhline(0, color='gray', linewidth=0.5)
ax.axvline(0, color='gray', linewidth=0.5)

# Plot the correlation circle
for i, feature in enumerate(features.columns):
    ax.arrow(0, 0, correlation_circle[i, 0], correlation_circle[i, 1],
             head_width=0.001, head_length=0.001, fc='b', ec='b')
    ax.text(correlation_circle[i, 0] * 1.5, correlation_circle[i, 1] * 1.5, feature,
            ha='center', va='center', fontsize=5, color='black')

# Add labels and grid
ax.set_xlabel('PC1', fontsize=5)
ax.set_ylabel('PC2', fontsize=5)
ax.grid(True, linestyle='--', linewidth=0.1)

# Adjust tick parameters
ax.tick_params(axis='both', which='both', length=0)

# Add a title
ax.set_title('Correlation Circle', fontsize=14, fontweight='bold')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix


# Define the number of classes
num_classes = 8

# Convert ordinal labels to integers
y_test = y_test.astype(int)
y_pred = y_pred.astype(int)

# Create confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=range(1, num_classes + 1))

# Calculate True Positives (TP)
TP = np.diag(cm)

# Calculate False Positives (FP)
FP = np.sum(cm, axis=0) - TP

# Calculate False Negatives (FN)
FN = np.sum(cm, axis=1) - TP

# Calculate True Negative (TN)
TN = np.sum(cm) - (TP + FP + FN)

# Calculate True Positive Rate (TPR)
TPR = TP / (TP + FN)

# Calculate False Positive Rate (FPR)
FPR = FP / (FP + TN)

# Calculate Precision
precision = TP / (TP + FP)

# Calculate Recall
recall = TP / (TP + FN)

# Calculate F1 Score
f1_score = 2 * (precision * recall) / (precision + recall)

# Print the calculated metrics
print("True Positives (TP):", TP)
print("False Positives (FP):", FP)
print("False Negatives (FN):", FN)
print("True Negative (TN):", TN)
print("True Positive Rate (TPR):", TPR)
print("False Positive Rate (FPR):", FPR)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```

```python
import pandas as pd

def drop_high_correlation_variables(data, threshold=0.75):
    # Compute the correlation matrix
    corr_matrix = data.corr().abs()

    # Create a mask to identify highly correlated features
    mask = corr_matrix > threshold

    # Find columns to drop
    columns_to_drop = []
    for col in mask.columns:
        correlated_cols = mask.index[mask[col]].tolist()
        columns_to_drop.extend(correlated_cols[1:])

    # Drop the highly correlated variables from the DataFrame
    data_filtered = data.drop(columns=set(columns_to_drop))

    return data_filtered
```
